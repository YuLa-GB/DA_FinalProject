```python
import requests
import pandas as pd
from datetime import datetime, timedelta
# Функция для получения вакансий
def get_vacancies(position, area='1', days=30):
    url = 'https://api.hh.ru/vacancies'
    params = {
        'text': position,
        'area': area,
        'date': f'1{days}',
        'per_page': 100,
        'page': 0
    }
    
    vacancies = []
    
    while True:
        response = requests.get(url, params=params)
        data = response.json()
        
        vacancies.extend(data['items'])
        
        if data['pages'] <= params['page'] + 1:
            break
        params['page'] += 1
    return vacancies
# Список специальностей
positions = ['Аналитик данных', 'Data Scientist', 'Data Engineer']
# Собираем вакансии
all_vacancies = []
for position in positions:
    all_vacancies.extend(get_vacancies(position))
# Конвертируем в DataFrame
df = pd.DataFrame(all_vacancies)
# Сохраняем в CSV файл
df.to_csv('vacancies.csv', index=False, encoding='utf-8')
print(f"Собрано {len(df)} вакансий.")
```

```python
import pandas as pd
# Загрузка DataFrame
df = pd.read_csv('final_vacancies_cleaned.csv')
# Столбцы, которые нужно раскрыть
columns_to_expand = ['department', 'area', 'salary', 'address', 'employer', 'schedule', 'experience', 'employment']
# Функция для обработки и раскрытия столбцов
def process_columns(df, column):
    if column in df.columns:
        # Разделяем значения на списки и удаляем пробелы
        df[column] = df[column].astype(str).str.split(',').apply(lambda x: [item.strip() for item in x])
        # Раскрываем массивы в строки
        return df.explode(column)
    return df
# Обработка всех указанных столбцов
for column in columns_to_expand:
    df = process_columns(df, column)
# Объединение значений в одну строку для каждого уникального индекса
df = df.groupby(df.index).agg(lambda x: ', '.join(x.astype(str).unique())).reset_index(drop=True)
# Удаление дубликатов, если необходимо
df = df.drop_duplicates()
# Сохранение обновленного DataFrame
df.to_csv('final_vacancies_cleaned_expanded.csv', index=False, encoding='utf-8')
print("Массивы в указанных столбцах были раскрыты и сохранены в новом CSV-файле.")
```

```python
import pandas as pd
import ast
# Загрузка DataFrame
df = pd.read_csv('final_vacancies_cleaned_expanded.csv')
# Функция для извлечения значения по ключу из строки словаря или списка словарей
def extract_value(dict_str, key):
    try:
        dict_obj = ast.literal_eval(dict_str)  # Преобразуем строку в объект
        if isinstance(dict_obj, dict):  # Если это словарь
            return dict_obj.get(key)  # Возвращаем значение по указанному ключу
        elif isinstance(dict_obj, list):  # Если это список
            # Извлечение значений по ключу из всех словарей в списке
            return [item.get(key) for item in dict_obj if isinstance(item, dict)]
        else:
            return None  # Возвращаем None для других типов
    except (ValueError, SyntaxError):
        return None  # Возвращаем None в случае ошибки
# Применение функции к столбцам с соответствующей обработкой
columns_to_extract = {
    'salary': 'from',  
    'area': 'name',
    'department': 'name',
    'type': 'name',
    'employer': 'name',
    'schedule': 'name',
    'professional_roles': 'name', 
    'experience': 'name',
    'employment': 'name',
}
for column, key in columns_to_extract.items():
    if column in df.columns:
        df[column] = df[column].apply(lambda x: extract_value(x, key))
# Преобразование salary в числовой тип
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')  # Преобразует, заменяя ошибки на NaN
# Обработать даты
try:
    df['published_at'] = pd.to_datetime(df['published_at'], errors='coerce')  # Заменяет ошибки на NaT
except Exception as e:
    print(f"Ошибка при преобразовании дат публикации: {e}")
# Преобразование в категориальный тип (например, для employment и experience)
categorical_columns = ['employment', 'experience']
for col in categorical_columns:
    try:
        if col in df.columns:
            df[col] = df[col].astype('category')
    except Exception as e:
        print(f"Ошибка при преобразовании столбца {col}: {e}")
# Проверка типов данных после преобразования
print(df.dtypes)
# Сохранение обновленного DataFrame (если требуется)
df.to_csv('final_vacancies_cleaned_expanded_converted.csv', index=False, encoding='utf-8')
print("Типы данных успешно преобразованы и DataFrame сохранен.")
```

1. На основе датафрейма 'cleaned_vacancies_with_dates.csv' построим модель, прогнозирующую зарплату для вакансий, у которых она не указана в столбце salary. 
Спрогнозируем зарплату как медиану на основе таких же вакансий и опыта работы.
Для вакансий без опыта работы оставим минимальное значение зарплаты из похожих вакансий.
Для оставшихся вакансий без зарплаты укажем значение зарплаты как медиана в зависимости от опыта.

```python

# 1. Загрузка данных
df = pd.read_csv('cleaned_vacancies_with_dates.csv')
# 2. Группировка данных для вычисления медианы зарплаты по комбинации name и experience
median_salary = df.groupby(['name', 'experience'])['salary'].median().reset_index()
median_salary.rename(columns={'salary': 'median_salary'}, inplace=True)
# 3. Группировка данных для вычисления минимальной зарплаты для вакансий без опыта
min_salary_no_experience = df[df['experience'] == 'Нет опыта'].groupby(['name'])['salary'].min().reset_index()
min_salary_no_experience.rename(columns={'salary': 'min_salary_no_experience'}, inplace=True)
# 4. Объединение с оригинальным DataFrame
df = df.merge(median_salary, on=['name', 'experience'], how='left')
df = df.merge(min_salary_no_experience, on=['name'], how='left')
# 5. Заполнение отсутствующих значений в столбце salary
df['salary'] = df['salary'].fillna(df['median_salary'])
# 6. Заполнение вакансий без опыта минимальными зарплатами
df.loc[df['experience'] == 'Нет опыта', 'salary'] = df.loc[df['experience'] == 'Нет опыта', 'salary'].fillna(df['min_salary_no_experience'])
# 7. Заполнение оставшихся вакансий без зарплаты медианой по опыту
median_salary_by_experience = df.groupby('experience')['salary'].median().reset_index()
median_salary_by_experience.rename(columns={'salary': 'median_salary_experience'}, inplace=True)
# 8. Объединить для заполнения по опыту
df = df.merge(median_salary_by_experience, on='experience', how='left')
# 9. Заполнить оставшиеся пропуски зарплаты медианой по опыту
df['salary'] = df['salary'].fillna(df['median_salary_experience'])
# 10. Удаление временных столбцов
df.drop(columns=['median_salary', 'min_salary_no_experience', 'median_salary_experience'], inplace=True)
# 11. Проверка результатов
print(df[['name', 'experience', 'salary']].head())
# 12. Сохранение обновленного DataFrame в CSV файл
df.to_csv('cleaned_vacancies_with_salary_predictions.csv', index=False)
```

4. 1. На основе датафрейма 'cleaned_vacancies_with_salary_predictions.csv' проведем исследование наиболее востребованных навыков и соотношение «хард» и «софт» навыков.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
# 1. Загрузка данных
df = pd.read_csv('cleaned_vacancies_with_salary_predictions.csv')
# 2. Определение ключевых навыков (как hard skills, так и soft skills) для цифровых профессий
key_skills = {
    'hard_skills': [
        'Python', 'Java', 'C++', 'SQL', 'JavaScript',
        'HTML', 'CSS', 'Machine Learning', 'Data Analysis',
        'Digital Marketing', 'Data Visualization', 'Cloud Computing', 'DevOps'
    ],
    'soft_skills': [
        'Командная работа', 'Коммуникация', 'Лидерство', 'Проблемное решение',
        'Управление временем', 'Креативность', 'Адаптивность', 'Критическое мышление',
        'Эмпатия', 'Конфликтология', 'Масштабируемость', 'Гибкость', 'Ответственность',
        'Самостоятельность', 'Аналитическое мышление'
    ]
}
# 3. Извлечение навыков из столбца 'snippet'
skills_list = []
for snippet in df['snippet'].dropna():
    for skill in key_skills['hard_skills'] + key_skills['soft_skills']:
        if skill.lower() in snippet.lower():  # Проверяем наличие навыка в тексте
            skills_list.append(skill)
# 4. Подсчет частоты навыков
skill_counts = Counter(skills_list)
# 5. Создание DataFrame из навыков и их частоты
skills_df = pd.DataFrame(skill_counts.items(), columns=['skill', 'count'])
# 6. Классификация навыков
skills_df['type_snip'] = skills_df['skill'].apply(
    lambda x: 'Hard Skill' if x in key_skills['hard_skills'] else ('Soft Skill' if x in key_skills['soft_skills'] else 'Other')
)
# 7. Подсчет количества хард и софт навыков
skills_summary = skills_df[skills_df['type_snip'].isin(['Hard Skill', 'Soft Skill'])].groupby('type_snip')['count'].sum().reset_index()
# 8. Проверка первых нескольких строк
sorted_skills_df = skills_df.sort_values(by='count', ascending=False)
print(sorted_skills_df.head(10))  # Выведет первые 10 строк с наибольшим количеством навыков
print(skills_summary)
# 9. Визуализация соотношения хард и софт навыков
plt.figure(figsize=(8, 6))
sns.barplot(data=skills_summary, x='type_snip', y='count', palette='viridis')
plt.title('Соотношение "Hard" и "Soft" навыков', fontsize=16)
plt.xlabel('Тип навыка', fontsize=14)
plt.ylabel('Количество навыков', fontsize=14)
plt.tight_layout()
plt.show()
```

4. 2. На основе датафрейма 'cleaned_vacancies_with_salary_predictions.csv' проведем исследование наиболее востребованных навыков: определим ключевые навыки (как hard skills, так и soft skills).
Интерпретация результатов:
После выполнения кода мы получили две столбчатые диаграммы, которые покажут:
- Как наличие hard skills влияет на среднюю зарплату.
- Как наличие soft skills влияет на среднюю зарплату.

Мы видим, что наличие soft skills увеличивает уровень зарплаты по сравнению с отсутствием этих навыков. Так же наличие как hard, так и soft навыков говорит о высокой значимости каждого типа навыков в цифровых профессиях.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
# 1. Загрузка данных
df = pd.read_csv('cleaned_vacancies_with_salary_predictions.csv')
# Удаляем строки с NaN в столбце salary и snippet
df = df.dropna(subset=['salary', 'snippet'])
# 2. Определение навыков
# Создание списков hard skills и soft skills 
hard_skills = ['Python', 'Java', 'JavaScript', 'SQL', 'HTML', 'CSS', 'Data Analysis', 'Machine Learning']
soft_skills = ['Командная работа', 'Коммуникация', 'Управление временем', 'Адаптивность', 'Критическое мышление']
# Функция для проверки наличия навыка в описании
def check_skills(snippet, skills):
    return any(skill.lower() in snippet.lower() for skill in skills)
# Поиск hard и soft skills в snipped и создание новых столбцов
df['has_hard_skills'] = df['snippet'].apply(lambda x: check_skills(x, hard_skills))
df['has_soft_skills'] = df['snippet'].apply(lambda x: check_skills(x, soft_skills))
# 3. Анализ влияния hard и soft навыков на зарплату
salary_hard_skills = df.groupby('has_hard_skills')['salary'].mean().reset_index()
salary_soft_skills = df.groupby('has_soft_skills')['salary'].mean().reset_index()
# 4. Визуализация результатов
plt.figure(figsize=(14, 6))
# График для hard skills
plt.subplot(1, 2, 1)
sns.barplot(data=salary_hard_skills, x='has_hard_skills', y='salary', palette='viridis')
plt.title('Влияние Hard Skills на Среднюю Зарплату', fontsize=16)
plt.xlabel('Наличие Hard Skills', fontsize=14)
plt.ylabel('Средняя Зарплата (руб.)', fontsize=14)
plt.xticks(ticks=[0, 1], labels=['Нет', 'Да'])
# График для soft skills
plt.subplot(1, 2, 2)
sns.barplot(data=salary_soft_skills, x='has_soft_skills', y='salary', palette='viridis')
plt.title('Влияние Soft Skills на Среднюю Зарплату', fontsize=16)
plt.xlabel('Наличие Soft Skills', fontsize=14)
plt.ylabel('Средняя Зарплата (руб.)', fontsize=14)
plt.xticks(ticks=[0, 1], labels=['Нет', 'Да'])
plt.tight_layout()
plt.show()
```
